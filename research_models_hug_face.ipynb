```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загружаем объединённый DataFrame, который мы создали ранее.
# Предположим, что у вас есть файл "compare_rubert_searubert.csv" с колонками:
# "Name", "Review", "Rating", "Sentiment_rubert", "Sentiment_searubert", "Match"
df_compare = pd.read_csv("compare_rubert_searubert.csv")

# --- 1. Гистограммы распределения сентимента для каждой модели ---

# Считаем количество примеров для каждого сентимента у каждой модели
sentiments_rubert = df_compare["Sentiment_rubert"].value_counts().sort_index()
sentiments_searubert = df_compare["Sentiment_searubert"].value_counts().sort_index()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x=sentiments_rubert.index, y=sentiments_rubert.values, palette="viridis")
plt.title("Распределение сентимента (rubert)")
plt.xlabel("Сентимент")
plt.ylabel("Количество примеров")

plt.subplot(1, 2, 2)
sns.barplot(x=sentiments_searubert.index, y=sentiments_searubert.values, palette="magma")
plt.title("Распределение сентимента (sea-rubert)")
plt.xlabel("Сентимент")
plt.ylabel("Количество примеров")

plt.tight_layout()
plt.show()

# --- 2. Тепловая карта (confusion matrix) для сравнения моделей ---

# Создаем таблицу сопряженности (confusion matrix) между предсказаниями двух моделей
conf_matrix = pd.crosstab(df_compare["Sentiment_rubert"], df_compare["Sentiment_searubert"])

plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.title("Матрица согласования сентиментов:\nrubert vs sea-rubert")
plt.xlabel("sea-rubert")
plt.ylabel("rubert")
plt.show()
```

```python
df_rubert = pd.read_csv("rubert_collected_data_with_sentiment.csv")
df_searubert = pd.read_csv("sea-rubert_collected_data_with_sentiment.csv")
```

```python
df_compare = pd.DataFrame({
    "Name": df_rubert["Name"],
    "Review": df_rubert["Review"],
    # "Rating": df_rubert["Rating"],
    "Sentiment_rubert": df_rubert["Sentiment"],
    "Sentiment_searubert": df_searubert["Sentiment"]
})
```

```python
df_compare["Match"] = (df_compare["Sentiment_rubert"] == df_compare["Sentiment_searubert"])
```

```python
df_compare[df_compare["Match"] == True] # Вывести первые 20 строк для наглядности
```

```python
match_rate = df_compare["Match"].mean()
print(f"Модели совпали в {match_rate * 100:.2f}% случаев")
```

```python
df_compare.to_csv("compare_rubert_searubert.csv", index=False, encoding="utf-8")
```
