## ФОРМУМ-ЖКХ

```python
forum_samolet_df = pd.read_csv("forum-zhk_samolet.csv")
forum_granel_df = pd.read_csv("forum-zhk_granel.csv")
```

```python
forum_samolet_df
```

```python
forum_samolet_df = forum_samolet_df.dropna()
forum_samolet_df = forum_samolet_df[~forum_samolet_df["Review"].str.startswith("Siteinfo: IP-адреса отзывов из одной", na=False)]
forum_samolet_df["Date"] = pd.to_datetime(forum_samolet_df["Date"], format="%d.%m.%Y %H:%M").dt.strftime("%Y-%m-%d")
forum_samolet_df["Rating"] = forum_samolet_df["Rating"].apply(int)
```

```python
forum_samolet_df["Developer"] = "samolet"
forum_samolet_df["Site"] = "forum_zkh"
```

```python
forum_granel_df = forum_granel_df.dropna()
forum_granel_df = forum_granel_df[~forum_granel_df["Review"].str.startswith("Siteinfo: IP-адреса отзывов из одной", na=False)]
forum_granel_df["Date"] = pd.to_datetime(forum_granel_df["Date"], format="%d.%m.%Y %H:%M").dt.strftime("%Y-%m-%d")
forum_granel_df["Rating"] = forum_granel_df["Rating"].apply(int)
```

```python
forum_granel_df["Developer"] = "granel"
forum_granel_df["Site"] = "forum_zkh"
```
