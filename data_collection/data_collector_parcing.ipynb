```python
import csv
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Настройка драйвера
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Открытие страницы
driver.get("https://novostroev.ru/zastroyshchiki/samolet-development/otzyvy/")

# Подождем, пока страница полностью загрузится
time.sleep(3)

# Создаем или открываем CSV файл для записи
with open('novostroev_reviews.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Записываем заголовки
    writer.writerow(['Name', 'Date', 'Review', 'Stars'])

    # Начинаем с первой страницы
    page = 1
    while True:
        # Парсинг данных
        reviews = driver.find_elements(By.CSS_SELECTOR, '.reviews-list__item')  # Находим все отзывы

        # Обрабатываем каждый отзыв
        for review in reviews:
            try:
                # Извлекаем имя автора
                author_name = review.find_element(By.CSS_SELECTOR, '.review-n-2__name').text.strip()
            except:
                author_name = None

            try:
                # Извлекаем дату
                review_date = review.find_element(By.CSS_SELECTOR, '.review-n-2__meta').text.strip()
            except:
                review_date = None

            try:
                # Извлекаем текст отзыва
                review_text = review.find_element(By.CSS_SELECTOR, '.review-n-2__text').text.strip()
            except:
                review_text = None

            try:
                # Извлекаем количество звезд
                stars_element = review.find_element(By.CSS_SELECTOR, '.stars__full')
                style_attribute = stars_element.get_attribute("style")
                
                # Извлекаем число процентов из style (например, "width: 20%")
                match = re.search(r'width:\s*(\d+)%', style_attribute)
                stars = int(match.group(1)) // 20 if match else None  # 20% = 1 звезда, 40% = 2 звезды и т.д.
            except:
                stars = None

            # Записываем данные в файл
            writer.writerow([author_name, review_date, review_text, stars])

        # Пытаемся перейти на следующую страницу
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, f'a[href*="page={page + 1}"]')
            next_button.click()
            page += 1
            time.sleep(3)  # Подождем, пока страница загрузится
        except Exception as e:
            print("Следующая страница не найдена или ошибка:", e)
            break

# Закрываем драйвер
driver.quit()
```
